<h1>NATIONALITY ANALYSIS</h1>

```python
Terrorists['country_txt'].unique()
```

```python
filtered_attacks = Terrorists[
    (Terrorists['country_txt'] == 'Serbia') &
    (Terrorists['iyear'] >= 2010)
]

attacks_df = filtered_attacks[['natlty1_txt', 'latitude', 'longitude']].dropna()

map_nationality = folium.Map(location=[44.0165, 21.0059], zoom_start=7)

HeatMap(data=attacks_df[['latitude', 'longitude']]).add_to(map_nationality)

map_nationality
```

<h1>REGIONAL ANALYSIS</h1>

```python
regional_attacks = Terrorists[
    (Terrorists['region_txt'] == 'Western Europe') &
    (Terrorists['iyear'] == 2017)
]

regional_df = regional_attacks[['latitude', 'longitude']].dropna()

map_regional = folium.Map(location=[50, 10], zoom_start=4)

HeatMap(data=regional_df[['latitude', 'longitude']]).add_to(map_regional)

map_regional
```

<h1>TARGET ANALYSIS</h1>

```python
target_attacks = Terrorists[
    (Terrorists['country_txt'] == 'New Zealand') &
    (Terrorists['iyear'] >= 2000)
]

target_df = target_attacks[['latitude', 'longitude']].dropna()

map_target = folium.Map(location=[-41, 174], zoom_start=5)

HeatMap(data=target_df[['latitude', 'longitude']]).add_to(map_target)

map_target
```

<h1>CASUALTIES ANALYSIS</h1>

```python
casualties_df = Terrorists[
    (Terrorists['iyear'] >= 2016) &
    (Terrorists['region_txt'] == 'North America') &
    (Terrorists['nkill'] + Terrorists['nwound'] > 0)
][['latitude', 'longitude', 'nkill', 'nwound']].dropna()

casualties_df['total_casualties'] = casualties_df['nkill'] + casualties_df['nwound']

casualties_df['normalized_casualties'] = (
    casualties_df['total_casualties'] - casualties_df['total_casualties'].min()) / (
    casualties_df['total_casualties'].max() - casualties_df['total_casualties'].min()
)

map_casualties = folium.Map(location=[40, -100], zoom_start=4)

heatmap_data = [[row['latitude'], row['longitude'], row['normalized_casualties']] for index, row in casualties_df.iterrows()]

HeatMap(heatmap_data).add_to(map_casualties)

map_casualties
```

<h1>ATTACK ANALYSIS</h1>

```python
attack_type_analysis = Terrorists[
    (Terrorists['region_txt'] == 'South Asia') &
    (Terrorists['iyear'] == 2016) &
    (Terrorists['attacktype1_txt'] == 'Armed Assault')
]

attack_type_df = attack_type_analysis[['latitude', 'longitude']].dropna()

map_attack_type = folium.Map(location=[22, 80], zoom_start=4.5)

HeatMap(data=attack_type_df[['latitude', 'longitude']]).add_to(map_attack_type)

map_attack_type
```
