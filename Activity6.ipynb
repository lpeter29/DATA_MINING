<h1>NATIONALITY ANALYSIS</h1>

```python
Terrorists['country_txt'].unique()
```

```python
filtered_attacks = Terrorists[
    (Terrorists['country_txt'] == 'Serbia') &
    (Terrorists['iyear'] >= 2010)
]

attacks_df = filtered_attacks.groupby(["latitude", "longitude"]).size().reset_index(name = 'Number of Cases')

map_nationality = folium.Map(location=[44.0165, 21.0059], zoom_start=7)

HeatMap(data=attacks_df[['latitude', 'longitude']]).add_to(map_nationality)

map_nationality
```

<h1>NATIONALITY INSIGHT</h1>

<h2>The only prone to terrrist in Serbia from 2010 onwards is on the southern and northern part of the Nation</h2>

<h1>REGIONAL ANALYSIS</h1>

```python
regional_attacks = Terrorists[
    (Terrorists['region_txt'] == 'Western Europe') &
    (Terrorists['iyear'] == 2017)
]

regional_df = regional_attacks.groupby(["latitude", "longitude"]).size().reset_index(name = 'Number of Cases')

map_regional = folium.Map(location=[50, 10], zoom_start=4)

HeatMap(data=regional_df[['latitude', 'longitude']]).add_to(map_regional)

map_regional
```

<h1>REGIONAL INSIGHT</h1>

<h2>The area prone to terrrist in Western Europe on the year 2017 is more on the western part and the middle part of the region</h2>

<h1>TARGET ANALYSIS</h1>

```python
Terrorists['targtype1_txt'].unique()
```

```python
target_attacks = Terrorists[(Terrorists['targtype1_txt'] == 'Government (Diplomatic)') &
    (Terrorists['country_txt'] == 'New Zealand') &
    (Terrorists['iyear'] >= 2000)
]

target_df = target_attacks.groupby(["latitude", "longitude"]).size().reset_index(name = 'Number of Cases')

map_target = folium.Map(location=[-41, 174], zoom_start=5)

HeatMap(data=target_df[['latitude', 'longitude']]).add_to(map_target)

map_target
```

<h1>TARGET INSIGHT</h1>

<h2>The area that targeted a Government in New Zealand on the year 2000 onwards is on the northern and middle part of the country</h2>

<h1>CASUALTIES ANALYSIS</h1>

```python
casualties_df = Terrorists[
    (Terrorists['iyear'] >= 2016) &
    (Terrorists['region_txt'] == 'North America') &
    (Terrorists['nkill'] + Terrorists['nwound'] > 0)
]

casualties_df = casualties_df.groupby(["latitude", "longitude"]).agg(
    total_casualties=("nkill", lambda x: x.sum() + casualties_df.loc[x.index, "nwound"].sum())
).reset_index()

casualties_df['normalized_casualties'] = (
    casualties_df['total_casualties'] - casualties_df['total_casualties'].min()) / (
    casualties_df['total_casualties'].max() - casualties_df['total_casualties'].min()
)

map_casualties = folium.Map(location=[40, -100], zoom_start=4)

HeatMap(data=casualties_df[['latitude', 'longitude', 'normalized_casualties']]).add_to(map_casualties)

map_casualties
```

<h1>CASUALTIES INSIGHT</h1>

<h2>The area that has the most casualties occured is on Las Vegas in the North America Region</h2>

<h1>ATTACK ANALYSIS</h1>

```python
attack_type_analysis = Terrorists[
    (Terrorists['region_txt'] == 'South Asia') &
    (Terrorists['iyear'] == 2016) &
    (Terrorists['attacktype1_txt'] == 'Armed Assault')
]

attack_type_df = attack_type_analysis.groupby(["latitude", "longitude"]).size().reset_index(name = 'Number of Cases')

map_attack_type = folium.Map(location=[22, 80], zoom_start=4.5)

HeatMap(data=attack_type_df[['latitude', 'longitude']]).add_to(map_attack_type)

map_attack_type
```

<h1>ATTACK INSIGHT</h1>

<h2>The highest concentration of armed assaults is observed in the western part of the South Asia Region from the year 2016</h2>
